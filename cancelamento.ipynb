```python
import pandas as pd 
import numpy as np

tabela = pd.read_csv("hotel_bookings.csv")
display(tabela)
```

```python
# identificando e removendo valores vazios 

display(tabela.info())
tabela = tabela.dropna()
display(tabela.info())
```

# vamos anÃ¡lisar quantos clientes cancelaram 



```python
display(tabela["is_canceled"].value_counts())
display(tabela["is_canceled"].value_counts(normalize=True).map("{:.1%}".format))

# 92.2% de nÃ£o cancelados e 7.8% de cancelados
```

```python
display(tabela["arrival_date_month"].value_counts(normalize=True))
display(tabela["arrival_date_month"].value_counts(normalize=True).map("{:.1%}".format))
# descobri que a maior parte do mes da data de chegada foi em novembro
```

```python
display(tabela.groupby("arrival_date_month").mean(numeric_only=True))
```

```python
# tirei a coluna abril por que dentre os meses que sepassaram foi o mÃªs que mais ouve numeros de cancelamentos
tabela = tabela[tabela["arrival_date_month"]!="april"]
display(tabela)
display(tabela["is_canceled"].value_counts())
display(tabela["is_canceled"].value_counts(normalize=True).map("{:.1%}".format))
```

```python
# Resort hotel teve mais hospedagens em comparaÃ§Ã£o com o hotel cidade
display(tabela["hotel"].value_counts(normalize=True).map("{:.1%}".format))
```

```python
display(tabela.groupby("hotel").mean(numeric_only=True))
```
