```python
import pandas as pd
import numpy as np
tabela = pd.read_csv("Netflix Userbase.csv")
display(tabela)

```

```python
# Removendo e identificando valores vazios

display(tabela.info())
tabela = tabela.dropna()
display(tabela.info)
```

```python
display(tabela["Subscription Type"].value_counts())
display(tabela["Subscription Type"].value_counts(normalize=True).map("{:.1%}".format))

# Pela analise feita nesse topico vemos que o plano bÃ¡sico Ã© o mais utilizado com 40%
```

```python
# Vamos analisar em que aparelho que as pessoas se conectÃ£o mais
display(tabela["Device"].value_counts(normalize=True).map("{:.1%}".format))
display(tabela["Device"].value_counts())

# Aparelho que que as pessoas se conectÃ£o mais Ã© o laptop 25.5%
```

```python
# Agora vamos analisar a nossa faixa etÃ¡ria 
display(tabela["Age"].value_counts(normalize=True).map("{:.1%}".format))
```

```python
display(tabela["Country"].value_counts(normalize=True).map("{:.1%}".format))
```

```python
display(tabela["Join Date"].value_counts(normalize=True).map("{:.1%}".format))
```

```python
display(tabela.groupby("Join Date").mean(numeric_only=True))
```

```python
display(tabela["Monthly Revenue"].value_counts(normalize=True).map("{:.1%}".format))
```
