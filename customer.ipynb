```python
import pandas as pd 
import numpy as np 

tabela = pd.read_csv("Telco-Customer-Churn.csv")
display(tabela)
```

```python
tabela.drop(["customerID"], axis=1)
```

```python
#1 Passo Vou análizar as colunas 
#MultipleLines = Se o cliente tem várias linhas ou não (Sim, Não, Sem atendimento telefônico)	

#InternetService = Provedor de serviços de internet do cliente (DSL, Fibra ótica, Não)

#OnlineSecurity	= Se o cliente tem segurança online ou não (Sim, Não, Sem serviço de internet)

#TechSupport = Suporte técnico

#streaming de TV e filmes

#Contract = contratos

#PaymentMethod = Forma de pagamento

#partner = sócio

#Chrun = é a coluna que mostra quantos clientes sairam 
```

```python
display(tabela.info())
tabela = tabela.dropna()
display(tabela.info())
```

```python
# Aqui vemos a porcentaegm de clientes que sairam 
display(tabela["Churn"].value_counts(normalize=True).map("{:.1%}".format))
```

```python
display(tabela["Contract"].value_counts(normalize=True).map("{:.1%}".format))
```

```python
# Analisando o contrato
display(tabela.groupby("Contract").mean(numeric_only=True))
# descobrimos aqui que a as cobranças mensais são maiores e a posse costuma ser menor 
```

```python
tabela = tabela[tabela["Contract"]!="Month-to-month"]
display(tabela)
display(tabela["Churn"].value_counts())
display(tabela["Churn"].value_counts(normalize=True).map("{:.1%}".format))

# Com isso já descobrimos que o contrato mensal é ruim 
```

```python
# Vamos análisar os serviços e tentar diminuir os numeros de clientes que sairam
```

```python
display(tabela.groupby("InternetService").mean(numeric_only=True))
display(tabela["InternetService"].value_counts(normalize=True).map("{:.1%}".format))


```

```python
display(tabela.groupby("OnlineSecurity").mean(numeric_only=True))
display(tabela["OnlineSecurity"].value_counts(normalize=True).map("{:.1%}".format))
```
