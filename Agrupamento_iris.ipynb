Agrupamentos com C-means

```python
import numpy as np 
import skfuzzy
from sklearn import datasets
from sklearn.metrics import confusion_matrix
```

#pip install -U scikit-fuzzy

```python
iris = datasets.load_iris()
```

Aqui nós criamos o cluster 

Aplicação do algoritimo definido três cluster (c = 3) e passando a matriz transposta (iris.data.T). Os outros parâmetros são os default indicados ma documentação 

iris.data = Você tem que passar a matriz transposta de iris.data usa-se o T para fazer essa matriz transposta 

O c= é o numero de cluster que você quer criar

```python
r = skfuzzy.cmeans(data= iris.data.T, c = 3, m = 2, error = 0.005, 
                   maxiter = 1000, init = None)
```

Obtendo as porcentagens de um registros pertencer a um cluster, que está na posição 1 da matriz retornada 

A variável r ela cria varias elementos 

```python
previsoes_porcentagem = r[1]
```

Visualização da probabilidade de um registro pertencer a cada um dos cluster (O somátorio é 1.0 que indica 100%)

```python
for x in range(150):
    print(previsoes_porcentagem[0][x], previsoes_porcentagem[1][x], previsoes_porcentagem[2][x])
```

O primeiro cluster estão entre 99 e 97 em com paração aos demais indo pro final os cluster ficam são tão claro

Geração de matriz de contigencia para comparação com as classe originais da base de dados

```python
previsoes = previsoes_porcentagem.argmax(axis=0)
resultado = confusion_matrix(iris.target, previsoes)
resultado
```

Na diagonal ele fez a observação certa mas tem 47, 37 errou
